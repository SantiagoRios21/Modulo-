Con el fin de entrenar diferentes modelos separamos los datos

```python
X_kntrain, X_kntest, y_kntrain, y_kntest = train_test_split(MatriX,y_label, test_size=0.3, random_state=21,stratify=y_label)

X_kntrain = np.array(X_kntrain)
X_kntest = np.array(X_kntest)
y_kntrain = np.array(y_kntrain)
y_kntest = np.array(y_kntest)

classes = pd.unique(y_label)
nns = np.arange(2,50,4)
```

```python
def cv_kn(nns,X_t=X_kntrain,y_t=y_kntrain,nsplit=4):

    
    ''' Esta función estima el numero de
        neighbors usando StratifiedKFold sobre 
        el conjunto de entrenamiento.
        para ello utiliza f1_score macro.
    '''
    average_acc = []
    average_f1 = []

    for n in nns:
        acc = []
        f1 = []

        for t_and_v_set in StratifiedKFold(n_splits=nsplit,random_state=21,shuffle=True).split(X_t,y_t):

            train, test = t_and_v_set
            knn = KNeighborsClassifier(n_neighbors=n,weights='distance',p=2)
            knn.fit(X_t[train],y_t[train])
            y_pred = knn.predict(X_t[test])

            acc.append(accuracy_score(y_t[test], y_pred))
            f1.append(f1_score(y_t[test], y_pred,average='macro'))


        average_acc.append(np.average(acc))
        average_f1.append(np.average(f1))


    average_acc = np.array(average_acc) 
    average_f1 = np.array(average_f1)

    return nns[np.where(average_f1==average_f1.max())[0][0]]
n_best = cv_kn(nns)
```

```python
%%time
n_k = n_best

def knn_classifier(n,X_knt=X_kntrain,y_knt=y_kntrain,X_knv=X_kntest,y_knv=y_kntest):
    '''
        Esta funcion utiliza KNeighbhors para un n dado y un
        set de entrenamiento/validación.
        Retorna un reporte con las metricas de cada clase, la matriz de confusión
        y las curvas ROC
    '''
    knn = KNeighborsClassifier(n_neighbors=n,weights='distance',p=2)
    knn.fit(X_knt,y_kntrain)
    y_pred = knn.predict(X_knv)
    
    print("Classification report for classifier %s:\n%s\n"
      % (knn, classification_report(y_knv, y_pred)))
    #Ploteamos
    cm = confusion_matrix(y_knv, y_pred,normalize='true',labels = classes)
    fig4, axss=plt.subplots(figsize=(20,17))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=classes)
    disp.plot(ax = axss)
    
    y_predn = y_pred.copy()
    y_knvn = y_pred.copy()
    
    #for i,ks in enumerate(classes):
        #y_predn = np.where(y_pred == str(ks) ,int(i), y_predn)
        #y_knvn = np.where(y_knv == str(ks) ,int(i), y_knvn)

    
    y_knvn = y_knvn.astype(int)
    y_predn = y_predn.astype(int)

    fig2, axss=plt.subplots(figsize=(12,8),nrows=1,ncols = 2)
    fig2.subplots_adjust(wspace=0.2,hspace=0.35)
    
    for i,axindex in enumerate(axss.flat):
        
        RocCurveDisplay.from_predictions(y_knvn, y_predn,pos_label=i,ax = axindex)
        axindex.set_title(classes[i])
        axindex.grid()
        
    
    return y_pred


y_predkn = knn_classifier(n_k)
```

```python
from sklearn.ensemble import RandomForestClassifier

#re definimos los set de entrenamiento y validacion con un conjunto de datos normalizado
X_rftrain, X_rftest, y_rftrain, y_rftest = train_test_split(MatriX,y_label, test_size=0.5, random_state=11,stratify=y_label)

X_rftrain = np.array(X_rftrain)
X_rftest = np.array(X_rftest)
y_rftrain = np.array(y_rftrain)
y_rftest = np.array(y_rftest)
ns = np.array([2**i for i in range(1,10)])
ns
```

```python
def cv_rf(ns,X_t=X_rftrain,y_t=y_rftrain,nsplit=2):
    
    '''
        Esta funcion estima el numero de arboles en 
        el clasificador, actua sobre el set de entrenamiento normalizado
    '''
    
    average_f1 = []


    for n in ns:
        acc = []
        f1 = []

        for t_and_v_set in StratifiedKFold(n_splits=nsplit,random_state=51,shuffle=True).split(X_t,y_t):
            train, test = t_and_v_set
            rf = RandomForestClassifier(n_estimators=n,criterion='gini',oob_score=True)
            rf.fit(X_t[train],y_t[train])
            y_pred = rf.predict(X_t[test])
            
            f1.append(f1_score(y_t[test], y_pred,average='macro'))
            
        average_f1.append(np.average(f1))
 
    average_f1 = np.array(average_f1)
    return ns[np.where(average_f1==average_f1.max())[0][0]]
```

```python
%%time
n_rfbest = cv_rf(ns)
```

```python
%%time
n_t = n_rfbest
def rf_classifier(n,X_rft=X_rftrain,y_rft=y_rftrain,X_rfv=X_rftest,y_rfv=y_rftest):
    
    '''Esta funcion utiliza Randomforest para un n dado y un
        set de entrenamiento/validación.
    '''
    rf = RandomForestClassifier(n_estimators=n,criterion='gini',oob_score=True)
    rf.fit(X_rft,y_rft)
    
    y_pred = rf.predict(X_rfv)
    
    print("Classification report for classifier %s:\n%s\n"
      % (rf, classification_report(y_rfv, y_pred)))
    
    #matriz de confusión
    
    cm = confusion_matrix(y_rfv, y_pred,labels = classes,normalize = 'true')
    fig, axs=plt.subplots(figsize=(20,17))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=classes)
    disp.plot(ax=axs)
    
    y_predn = y_pred.copy()
    y_rfvn = y_pred.copy()
    
    #roccurve
    for i,ks in enumerate(classes):
        y_predn = np.where(y_pred == ks ,int(i), y_predn)
        y_rfvn = np.where(y_rfv == ks ,int(i), y_rfvn)

        
    
    y_rfvn = y_rfvn.astype(int)
    y_predn = y_predn.astype(int)

    fig2, axss=plt.subplots(figsize=(23,15),nrows=1,ncols = 2)
    fig2.subplots_adjust(wspace=0.2,hspace=0.35)
    
    for axindex,i in zip(axss.flat,range(len(classes)+1)):
        RocCurveDisplay.from_predictions(y_rfvn, y_predn,pos_label=i,ax = axindex)
        axindex.set_title(classes[i])
        axindex.grid()
    
    return y_pred

y_predrf = rf_classifier(n_t)
```

```python
def cv_MLP(alphas,X_t=X_kntrain,y_t=y_kntrain,nsplit=4):

    
    ''' Esta función estima el numero de
        neighbors usando StratifiedKFold sobre 
        el conjunto de entrenamiento.
        para ello utiliza f1_score macro.
    '''
    average_acc = []
    average_f1 = []

    for a in alphas:
        acc = []
        f1 = []

        for t_and_v_set in StratifiedKFold(n_splits=nsplit,random_state=21,shuffle=True).split(X_t,y_t):

            train, test = t_and_v_set
            MLP = MLPClassifier(alpha=a)
            MLP.fit(X_t[train],y_t[train])
            y_pred = MLP.predict(X_t[test])

            acc.append(accuracy_score(y_t[test], y_pred))
            f1.append(f1_score(y_t[test], y_pred,average='macro'))


        average_acc.append(np.average(acc))
        average_f1.append(np.average(f1))


    average_acc = np.array(average_acc) 
    average_f1 = np.array(average_f1)

    return alphas[np.where(average_f1==average_f1.max())[0][0]]
```
